# 0 . End-to-end sci-kit-learn workflow

```python
# Get the prepared data ready. (heart disease data in our case)
import pandas as pd
import numpy as np
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
# I want to use the other columns to predict what the target value will be as the output
# so values from the all the other columns will serve as my input parameter called x and y will be the
# target value from target column after our operation on the model which is the function or f(x)
#hence; x is he feature matrix and y is the label

x = heart_disease.drop("target", axis=1) #this drops the target column since we dont want it here
x
```

```python
y = heart_disease["target"]
y
```

```python
# chose the right model and hyperparameters; hyperparameters are parameters that can be 
#changed to improve or affect the model
# we are using RandomForestClassifier model in our case
from sklearn.ensemble import RandomForestClassifier
#define the modelwith model or clf as name
model = RandomForestClassifier()

#choose hyperparameters I want for my model
# I used the default hyperparameter for this project
model.get_params()
```

```python
# I fit the model to the dataset by splitting the feature matrix , x and label, y into test and train data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5)
#this split the x and y data classifications into train and test and pass them into the model/function with
#50% test size which means 50$ of the data is for training

#now lets fit
model.fit(x_train, y_train)
#wow! my model has been fitted to the data

```

```python
# I want to evaluate the module on training and test data, like checking how accuarate the model performs
model.score(x_train, y_train)
# 1.0 is the result, that is, it is 100% accurate
```

```python
# But evaluating the test data, should not expect an 100% because the model wasnt trained with it
#lets see
model.score(x_test, y_test)
#huge!! fanstastic, though not 100 but very high 87%
```

```python
#mind you, asides accuracy test using the model.score() method, there are other ways of evaluation
#classification_report, confusion_matrix, accuracy score
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds )
```

```python
accuracy_score(y_test, y_preds)
```

```python
#I want to improve my model
# for example trying different amount of n_extimators instead of the default 100
#lets seed the array to fix set
np.random.seed(30)
for i in range(5, 100, 5): #group estimators in 5s up to 100
    print(f"Trying model with {i} estimators...")
    model = RandomForestClassifier(n_estimators=i).fit(x_train, y_train)
    print(f"Model accuracy on test is : {model.score(x_test, y_test) * 100:.2f}% accurate")
    print("")

```

# filling missing data with scikitlearn

```python
# i have to reread the car_sales_missing data extended
car_sales_missing = pd.read_csv("car-sales-extended-missing-data.csv")
car_sales_missing
```

```python
car_sales_missing.isna().sum() #alright the NAN rows are back, see below
```

```python
#remove NaN from the price column
car_sales_missing.dropna(subset = ["Price"], inplace=True)
car_sales_missing.isna().sum()
```

```python
X = car_sales_missing.drop("Price", axis=1)
y = car_sales_missing["Price"]
X.isna().sum()
```

```python
y.isna().sum()
```

```python
# now i fill in missing data cells with scikitlearn
# i import the SimpleInputer and ColumnTransformer modules
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

catg_imputer = SimpleImputer(strategy="constant", fill_value="missing")
door_imputer = SimpleImputer(strategy="constant", fill_value=4) #4 is like the mean of the number of doors
num_imputer = SimpleImputer(strategy= "mean")

#define the features
catg_features = ["Make", "Colour"]
door_features = ["Doors"]
num_features = ["Odometer (KM)"]

imputer = ColumnTransformer([("catg_imputer", catg_imputer, catg_features),
                              ("door_imputer", door_imputer, door_features),
                              ("num_imputer", num_imputer, num_features)])
filled_X = imputer.fit_transform(X)
filled_X



```

```python
#fantastic, it has worked
car_sales_filled = pd.DataFrame(filled_X, columns=["Make", "Colour", "Door", "Odometer (KM)"])
car_sales_filled
```

```python
car_sales_filled.isna().sum() # NO MORE MISSING VALUES
```

```python
# its time to convert all strings to integer (binary numbers)
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
one_hot = OneHotEncoder()
cat_features = ["Make", "Colour", "Door"]
transformer = ColumnTransformer([("one_hot", 
                                  one_hot, cat_features)], 
                                remainder="passthrough")

transformed_X_filled = transformer.fit_transform(car_sales_filled)
transformed_X_filled
```

```python
#beautiful, I got the car_sales_missing data filled and converted the strings to numbers
#its time to model it, so i will import my choice of model and splitter
np.random.seed(42)
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
model = RandomForestRegressor()
X_train, y_train, X_test, y_test = train_test_split(transformed_X_filled, y, test_size=0.2)
model.fit(X_train, y_train)
```
